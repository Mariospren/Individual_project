## Removing outliers




After identifying the features with outlier issues, we'll take a closer look by visualizing them and then applying a function

```python

plt.figure(figsize=(15, 15))

plt.subplot(3, 1, 1)
plt.hist(df['price'])
plt.title('Price Histogram')
plt.xlabel('Price')
plt.ylabel('Count')

plt.subplot(3, 1, 2)
plt.hist(df['year'])
plt.title('Year Histogram')
plt.xlabel('Year')
plt.ylabel('Count')

plt.subplot(3, 1, 3)
plt.hist(df['odometer'])
plt.title('Odometer Histogram')
plt.xlabel('Kilometers')
plt.ylabel('Count')

plt.show()

```

Making a function in order to remove outliers



```python
def handle_outliers(df, column_name):
    """
    Removes outliers from a specific column in a pandas DataFrame.

    Parameters:
        df (pandas.DataFrame): The pandas DataFrame containing the data.
        column_name (str): The name of the column from which outliers will be removed.

    Returns:
        pandas.DataFrame: A DataFrame with outliers removed from the specified column.
    """
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df = df[(df[column_name] >= lower_bound) & (df[column_name] <= upper_bound)]
    return df
```

```python
# Calculate the number of rows before removing outliers
original_rows = len(df)

# Remove outliers from the specified columns
df = handle_outliers(df, 'price')
df = handle_outliers(df, 'year')
df = handle_outliers(df, 'odometer')




# Calculate the number of rows after removing outliers
rows_after = len(df)





# Calculate the percentage of rows removed
percentage_removed = ((original_rows - rows_after) / original_rows) * 100

print(f"Percentage of rows removed: {percentage_removed:.2f}%")
```

We have removed 7% of the rows from the dataframe. It shouldn't affect our analysis with the rest of the dataframe

# ML process

I want to use ML to assess the importance of various features of interest, such as year, odometer, and condition. I chose the "Ridge" linear regression because it best fit the data.


### Preprocessing

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn import model_selection 
from sklearn.preprocessing import StandardScaler


```

Lets pick the revelnt variables for our ML process

```python
df_ml = df[['price', 'year', 'condition', 'odometer']]
```

```python
df_ml
```

Turning the categorical variables into numerical variables




```python

df_onehot = pd.get_dummies(df_ml, columns=['condition'])
df_onehot.head()


```

```python


# Initialize the StandardScaler object to normalize the data
scaler = StandardScaler()

df_onehot['price'] = scaler.fit_transform(df_onehot[['price']])
df_onehot['year'] = scaler.fit_transform(df_onehot[['year']])
df_onehot['odometer'] = scaler.fit_transform(df_onehot[['odometer']])
df_onehot['condition_excellent'] = scaler.fit_transform(df_onehot[['condition_excellent']])
df_onehot['condition_good'] = scaler.fit_transform(df_onehot[['condition_good']])
df_onehot['condition_new'] = scaler.fit_transform(df_onehot[['condition_new']])
df_onehot['condition_salvage'] = scaler.fit_transform(df_onehot[['condition_salvage']])

```

We can see the result of the normalization




```python
df_onehot
```

```python
# Separating the features (x) and the target variable (y). 
# 'price' is the target variable, so it is dropped from the features set.
x, y = df_onehot.drop(columns=['price']), df_onehot['price']

# Creating a list of feature names by removing the target variable 'price' from the column names.
features_cars = list(df_onehot.columns)
features_cars.remove('price')
```

### Ridge Regression






```python
model = Ridge()

```

```python
# Fit the model using the features (x) and the target variable (y)
model.fit(x, y)

print(f'model intercept: {model.intercept_}')
print(f'model coefficients: {model.coef_}')

# Plotting the Ridge Regression Coefficients
plt.xticks(range(len(features_cars)), features_cars, rotation=45)
plt.title('Ridge Regression Coefficients')
_ = plt.plot(model.coef_, '-', model.coef_, 'r^')


```

### Conclusion
